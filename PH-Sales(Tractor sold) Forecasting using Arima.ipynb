```python
from statsmodels.tsa.stattools import adfuller
#augmented dicky fuller 
adfuller(ts_sales)
```

```python
def test_stationary(timeseries):
    roll_mean =timeseries.rolling(12).mean()
    roll_std = timeseries.rolling(12).std()
    
    plt.figure(figsize=(18,10))
    plt.grid("both")
    plt.plot(timeseries)
    plt.plot(roll_mean,label = "rol_mean")
    plt.plot(roll_std,label = "rol_std")
    plt.legend()
    print("--"*20)
    print("Augmented dicky fuller Test:")
    dftest = adfuller(timeseries,autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=["Test-stats","p-value","#lag-used","No of observation"])
    for key , value in dftest[4].items():
        dfoutput["confidense interval (%s)"%key] = value
    print(dfoutput)
```

```python
test_stationary(ts_sales)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# ARIMA
#p,d,q
#p = auto regression
#d = differencing
#q = moving average
```

```python
def arima_model(X,y,arima_order):
    history=[i for i in X]
    prediction=list()
    for i in range(len(y)):
        model = ARIMA(history, order=arima_order)
        model_fit = model.fit()
        y_hat = model_fit.forecast()
        prediction.append(y_hat)
        history.append(y[i])
    mse = np.sqrt(mean_squared_error(y, prediction))
    return mse
```

```python
def evalute_arima_model(dataset, test,p_value, d_value,q_value):
    dataset = dataset.astype("float32")
    best_score, best_cfg = float('inf'), None
    for i in p_value:
        for j in d_value:
            for k in q_value:
                order = (i,j,k)
                rmse = arima_model(dataset, test,order)
                try:
                    if rmse < best_score:
                        best_score, best_cfg = rmse, order
                        print(rmse, order)
                except:
                    continue
    print(f"best scor and best cfg : {best_score} and {best_cfg}")          
                    
    
```
